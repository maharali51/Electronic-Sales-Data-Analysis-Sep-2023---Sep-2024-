### **Univariate Analysis of "Age" Using a Histogram Chart**

```python
# Univariate Analysis for 'Age'

plt.hist(df['Age'], bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Frequency')



# Find the center of each bin
counts, bins, patches =plt.hist(df['Age'], bins=10, color='blue', edgecolor='black')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
for count, x in zip(counts, bin_centers):
    plt.text(x, count, str(int(count)), ha='center', va='bottom')
```

**Univariate Analysis of "Time-based - Sales over Time" Using a Line Chart**

```python
# Univariate Analysis Time-based - Sales over Time

# Convert 'Purchase Date' to datetime format
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

# Group by Purchase Date and count sales
sales_over_time = df.groupby(df['Purchase Date'].dt.to_period('M')).size()
```

```python
# Line chart for sales over time

sales_over_time.plot(kind='line', marker='o', color='orange')
plt.title('Sales Over Time (Monthly)')
plt.xlabel('Month')
plt.ylabel('Number of Sales')

# If names are overlapping i am using this code only for practice
plt.xticks(rotation=-20, ha='right')

```

# **3-TIME SERIES ANALYSIS**

```python
# Time Series Analysis:

import pandas as pd
import matplotlib.pyplot as plt
path = 'Electronic_sales_Sep2023-Sep2024.csv'
# df = dataframe'
df =pd.read_csv(path)

print(df.columns)

```

```python
# Convert 'Purchase Date' to datetime format

df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
```

```python
# Set 'Date' as the index for time series analysis

df.set_index('Purchase Date', inplace=True)
```

```python
# Now proceed with the time series analysis (resampling, plotting)

monthly_sales = df['Total Price'].resample('M').sum()
```

**Plot Time Series Data  Using a Chart**

```python
# Plot the time series data
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
monthly_sales.plot()
plt.title('Monthly Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45 , ha='right')

```
